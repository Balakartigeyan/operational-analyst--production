```python
import pandas as pd
import numpy as np

print("Quantifying energy waste due to equipment stress...")

# --- Re-create df_process and Operational_State for robustness ---
# Load the CSV file into a DataFrame if not already loaded (assuming it might be a fresh run or kernel restart)
df_process = pd.read_csv('/content/ALL Process.csv')

# Convert 'Time' column to datetime objects, handling mixed formats and errors
df_process['Time'] = pd.to_datetime(df_process['Time'], errors='coerce', format='mixed')

# Set 'Time' column as the DataFrame's index
df_process.set_index('Time', inplace=True)

# Dynamically identify data streams (copied from earlier steps)
unique_sources = df_process['Data Source'].unique()

import_flow_tag = None
for source in unique_sources:
    if 'TR1010ML11_CO1_FIT' in source and 'pisystem-malaysia' in source:
        import_flow_tag = source
        break

export_flow_tag = None
for source in unique_sources:
    if 'TR1050ML11_CO1_FIT' in source and 'pisystem-malaysia' in source:
        export_flow_tag = source
        break

blending_tags = [tag for tag in unique_sources if '_QQIT' in tag or '_WQIT' in tag]
motor_current_tags_all = [tag for tag in unique_sources if '|Current' in tag]

# Filter DataFrames to calculate thresholds (copied from earlier steps)
df_import_temp = df_process[df_process['Data Source'] == import_flow_tag]
df_export_temp = df_process[df_process['Data Source'] == export_flow_tag]
df_blending_temp = df_process[df_process['Data Source'].isin(blending_tags)]
df_motor_current_temp = df_process[df_process['Data Source'].isin(motor_current_tags_all)]

# Define thresholds based on descriptive statistics (copied from earlier steps)
import_threshold = df_import_temp['Value'].mean() * 0.1
if pd.isna(import_threshold) or import_threshold == 0: import_threshold = 100
export_threshold = df_export_temp['Value'].mean() * 0.1
if pd.isna(export_threshold) or export_threshold == 0: export_threshold = 100
# ADJUSTED: Set blending_threshold based on user feedback (max normal rate of 10k-10.5k)
blending_threshold = 10500.0 # Set to the upper end of the user's stated normal maximum rate

motor_current_threshold = df_motor_current_temp['Value'].min() + 1
if pd.isna(motor_current_threshold) or motor_current_threshold < 1: motor_current_threshold = 5

# Define operational state function (copied from earlier steps)
def define_operational_state(row):
    source = row['Data Source']
    value = row['Value']

    if 'TR1010ML11_CO1_FIT' in source and 'pisystem-malaysia' in source:
        if value > import_threshold: return 'Producing/Running'
        elif value > 0 and value <= import_threshold: return 'Idle/Low Flow'
        else: return 'Downtime'
    elif 'TR1050ML11_CO1_FIT' in source and 'pisystem-malaysia' in source:
        if value > export_threshold: return 'Producing/Running'
        elif value > 0 and value <= export_threshold: return 'Idle/Low Flow'
        else: return 'Downtime'
    elif '_QQIT' in source or '_WQIT' in source:
        if value > blending_threshold: return 'Producing/Running'
        elif value > 0 and value <= blending_threshold: return 'Idle/Low Flow'
        else: return 'Downtime'
    elif '|Current' in source:
        if value > motor_current_threshold: return 'Running'
        elif value > 0 and value <= motor_current_threshold: return 'Idle'
        else: return 'Downtime'
    return 'Unknown'

# Apply the function to create a new 'Operational_State' column
df_process['Operational_State'] = df_process.apply(define_operational_state, axis=1)
# --- End of re-creation of df_process and Operational_State ---


# Assuming typical industrial motor parameters (these are placeholders and should be confirmed)
VOLTAGE = 400   # Volts (e.g., for a 3-phase industrial motor)
POWER_FACTOR = 0.8 # Typical power factor for induction motors
# Assuming 3-phase motor for sqrt(3)

# Re-define df_motor_current, df_motor_running, stress_threshold, and df_stressed_periods for robustness
motor_current_tags = [tag for tag in df_process['Data Source'].unique() if '|Current' in tag]
df_motor_current = df_process[df_process['Data Source'].isin(motor_current_tags)]
df_motor_running = df_motor_current[df_motor_current['Operational_State'] == 'Running']

# Calculate mean and standard deviation of current during 'Running' state
running_current_mean = df_motor_running['Value'].mean()
running_current_std = df_motor_running['Value'].std()

# Define a stress threshold (e.g., 2 times the standard deviation above the mean)
stress_threshold = running_current_mean + (2 * running_current_std)

df_stressed_periods = df_motor_current[
    (df_motor_current['Operational_State'] == 'Running') &
    (df_motor_current['Value'] > stress_threshold)
]

# Recalculate avg_stressed_current and running_current_mean from the re-defined dataframes
avg_stressed_current = df_stressed_periods['Value'].mean()
# running_current_mean is already calculated above

interval_seconds = 5 # Re-defining for robustness, assuming 5 seconds per data point

# Calculate power during stressed periods (in kW)
# Power (kW) = (Voltage * Current * Power Factor * sqrt(3)) / 1000
power_stressed_kw = (VOLTAGE * avg_stressed_current * POWER_FACTOR * (3**0.5)) / 1000

# Calculate power during normal running periods (in kW)
power_normal_kw = (VOLTAGE * running_current_mean * POWER_FACTOR * (3**0.5)) / 1000

# Duration of each data point in hours
interval_hours = interval_seconds / 3600

# Total duration of stressed periods in hours
total_stressed_hours = len(df_stressed_periods) * interval_hours

# Calculate wasted energy (additional energy consumed due to stress) in kWh
wasted_energy_kwh = (power_stressed_kw - power_normal_kw) * total_stressed_hours

print(f"\nAssumed Voltage: {VOLTAGE} V")
print(f"Assumed Power Factor: {POWER_FACTOR}")
print(f"Average current during stressed periods: {avg_stressed_current:.2f} A")
print(f"Mean current during normal running periods: {running_current_mean:.2f} A")
print(f"Average power during stressed periods: {power_stressed_kw:.2f} kW")
print(f"Average power during normal running periods: {power_normal_kw:.2f} kW")
print(f"Total duration of stressed periods: {total_stressed_hours:.2f} hours")
print(f"Quantified wasted energy due to stress: {wasted_energy_kwh:.2f} kWh")

print("Energy waste quantification complete.")
```
